```python
import numpy as np

np.random.seed(42)

array = np.random.uniform(0, 10, size=20)
print("Original Array:")
print(array)

array_rounded = np.round(array, 2)
print("\nArray rounded to 2 decimal places:")
print(array_rounded)
```

```python
minimum = np.min(array_rounded)
maximum = np.max(array_rounded)
median = np.median(array_rounded)
print("Minimum:", minimum)
print("Maximum:", maximum)
print("Median:", median)
```

```python
array_modified = array_rounded.copy()
mask = array_modified < 5
array_modified[mask] = array_modified[mask] ** 2
print(array_modified)
```

```python
def numpy_alternate_sort(array):

    if array.size == 0:
        return array.copy()

    sorted_array = np.sort(array)

    result = np.zeros_like(sorted_array)

    left = 0
    right = len(sorted_array) - 1
    for i in range(len(sorted_array)):
        if i % 2 == 0:
            result[i] = sorted_array[left]
            left += 1
        else:
            result[i] = sorted_array[right]
            right -= 1

    return result
```

```python
numpy_alternate_sort(array_modified)
```
