```python
import pandas as pd
import numpy as np

np.random.seed(42)

names = ['Alice', 'Bob', 'Charlie', 'David', 'Emma', 'Frank', 'Grace', 'Hannah', 'Ian', 'Julia']
subjects = ['Math', 'Science', 'English', 'History', 'Computer Science']

student_indices = np.random.randint(0, len(names), size=10)
subject_indices = np.random.randint(0, len(subjects), size=10)
scores = np.random.randint(50, 101, size=10)

df = pd.DataFrame({
    'Name': [names[i] for i in student_indices],
    'Subject': [subjects[i] for i in subject_indices],
    'Score': scores,
    'Grade': ''
})
```

```python
def assign_grade(score):
    if score >= 90:
        return 'A'
    elif score >= 80:
        return 'B'
    elif score >= 70:
        return 'C'
    elif score >= 60:
        return 'D'
    else:
        return 'F'

df['Grade'] = df['Score'].apply(assign_grade)
```

```python
def pandas_filter_pass(dataframe):

    return dataframe[dataframe['Grade'].isin(['A', 'B'])]
```
