```python
import random
import pandas as pd

positive_templates = [
    "An exhilarating journey from start to finish.",
    "A heartwarming story with brilliant performances.",
    "A breathtaking masterpiece that left me speechless.",
    "An outstanding film with a compelling plot.",
    "A delightful experience; I highly recommend it.",
    "Love it! A beautiful masterpiece; Surely watching again",
    "Highly recomended! Killed it! Marvelous",
]

negative_templates = [
    "A tedious and uninspired attempt at storytelling.",
    "A disappointing movie with lackluster acting.",
    "A confusing plot that goes nowhere.",
    "An overhyped film that fails to deliver.",
    "A forgettable experience; I would not watch again.",
    "An absolute diasspointment! Waste of time",
    "A worthless experinece and highly not recommended",
]

positive_reviews = random.choices(positive_templates, k=50)
negative_reviews = random.choices(negative_templates, k=50)

reviews = positive_reviews + negative_reviews
sentiments = ['positive'] * 50 + ['negative'] * 50

df = pd.DataFrame({
    'Review': reviews,
    'Sentiment': sentiments
})
df = df.sample(frac=1, random_state=42).reset_index(drop=True)

print(df.head())
```

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer(max_features=500, stop_words='english')
X = vectorizer.fit_transform(df['Review'])
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, df['Sentiment'], test_size=0.2, random_state=42
)
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score

model = MultinomialNB()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f"Test Set Accuracy: {accuracy:.2f}")
```

```python
def predict_review_sentiment(model, vectorizer, review):

    features = vectorizer.transform([review])
    return model.predict(features)[0]
```

```python
sample_review = "An exhilarating journey with breathtaking visuals."
print(predict_review_sentiment(model, vectorizer, sample_review))
```
